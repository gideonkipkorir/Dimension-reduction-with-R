## t- distribution Stocastic

```r
# Loading our tnse library
# 
library(Rtsne)
```

```r
#For plotting
colors = rainbow(length(unique(spm_num$Product.line)))
names(colors) = unique(spm_num$Product.line)
```

```r
#Executing the algorithm on curated data

tsne <- Rtsne(spm_num[,-1], dims = 2, perplexity=30, verbose=TRUE, max_iter = 500)

# Getting the duration of execution
# 
exeTimeTsne <- system.time(Rtsne(spm_num[,-1], dims = 2, perplexity=30, verbose=TRUE, max_iter = 500))
```

```r
plot(tsne$Y, t='n', main="tsne")
text(tsne$Y, labels=spm_num$Product.line, col=colors[spm_num$Product.line])
```

# Feature Selection methods
## 1: Filter Methods 

```r
library(caret)
library(corrplot)
#ensuring all our datasets are numeric
str(spm_num)
```

```r
#transforming product.line feature into numeric
spm_num$Payment <- as.numeric(spm_num$Payment)
str(spm_num)
```

```r
# Calculating the correlation matrix
correlationMatrix <- cor(spm_num)
```

```r
#Find attributes that are highly correlated
highlyCorrelated <- findCorrelation(correlationMatrix, cutoff=0.75)
```

```r
# Highly correlated attributes

highlyCorrelated

names(spm_num[,highlyCorrelated])
```

There are features that were found to be highly correlated. Hence will be dropped.

```r
# We can remove the variables with a higher correlation 
# and comparing the results graphically as shown below
# ---
# 
# Removing Redundant Features 
# ---
# 
spm_num2<-spm_num[-highlyCorrelated]

# Performing our graphical comparison
# ---
# 
par(mfrow = c(1, 2))
corrplot(correlationMatrix, order = "hclust")
corrplot(cor(spm_num2), order = "hclust")
```

## 2: Wrapper Methods 

```r
library(clustvarsel)
library(mclust)
```

```r
# Sequential forward greedy search (default)
# ---
#
out = clustvarsel(spm_num, G = 1:7)
out
```

From the above results, if we had a clustering problem, product.line and Branch would gave been the ideal features for clustering and the rest of the features will be left out.

```r
Subset1 = spm_num[,out$subset]
mod = Mclust(Subset1, G = 1:7)
summary(mod)
```

```r
plot(mod,c("classification"))
```
